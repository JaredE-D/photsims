```python
import numpy as np 
import matplotlib.pyplot as plt
import csv
```

```python
data_array = []
with open("sps_bo_log_20251121-054455.csv", 'r', newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader, None)  # Skip the header row
    for row in reader:
        data_array.append(row)
```

```python
DAT = np.array(data_array).T
DAT = DAT.astype(float)
```

```python
plt.plot(DAT[-1])
```

```python
def get_prev_results(csvname):
    data = np.loadtxt(csvname, skiprows=1,delimiter=',', dtype=float, encoding='utf-8')
    x0 = [list(row[0:4]) for row in data]
    y0 = [row[4] for row in data]
    return x0, y0

x1, y1 = get_prev_results("../results_bayesoptQ/sps_bo_log_20251129-030332.csv")
```
